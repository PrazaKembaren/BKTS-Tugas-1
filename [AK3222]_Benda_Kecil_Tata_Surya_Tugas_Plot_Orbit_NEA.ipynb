```python
import pandas as pd
import matplotlib.pyplot as plt
import rebound
```

```python
df = "https://raw.githubusercontent.com/APS-ITERA/komputasi/main/dataviz/nea.csv" #url to load the data
df = pd.read_csv(df)
df
```

```python
df['p'] = df['a']*(1-df['e'])
print(df[df['p'] <= 0.47])
print(len(df[df['p'] <= 0.47]))
```

```python
df = df[["full_name", "a", "e", "i", "w"]]  #sorting the column
df = df.loc[1801:1900]                      #sorting rows (100) depending on nomor absen
df.to_csv("data.csv")
df
```

```python
#redefine orbital elements
a = df['a']
e = df['e']
inc = df['i']
om = df['w']
objek = df['full_name']
```

```python
#simulation
sim = rebound.Simulation()
sim.add(["Sun","Mercury","Venus","399","Mars"])
sim.save_to_file("ss.bin")
```

```python
#start simulation for the asteroid
sim2 = rebound.Simulation()
sun = sim.particles[0]
sun = sim2.add(sun)
```

```python
sim = rebound.Simulation("ss.bin")

for i in range(1801,1900):
  sim2.add(a=a[i], e=e[i], inc=inc[i], omega=om[i], primary=sun)
#sim3 = rebound.Simulation()
#moon = sim.particles[4]
#sim3.add(moon)
```

```python
from rebound.simulation import Particles
```

```python
#Plotting
sim.move_to_com()
ob1 = rebound.OrbitPlot(sim, orbit_style="solid", lw=1)
sim2.move_to_com()
ob2 = rebound.OrbitPlot(sim2, orbit_style = None,  fig=ob1.fig, ax = ob1.ax)
ob1.orbits[0].set_linestyle("--")
ob1.orbits[1].set_linestyle("--")
ob1.orbits[2].set_linestyle("--")
ob1.orbits[3].set_linestyle("--")
ob1.particles.set_sizes([1.631039e-5*100000, 4.04484e-5*100000, 4.26343e-5*100000, 2.27075e-5*100000])   #rencananya pake actual size by radius planet

ob1.particles.set_color(["blue"])
ob2.particles.set_sizes([6])
ob1.particles.set_sizes([40])
#ob3 = rebound.OrbitPlot(sim3, primary=sim.particles[3], show_primary=False, fig=ob1.fig, ax = ob1.ax)
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.ylabel("y [AU]")
plt.xlabel("x [AU]")
plt.title("Orbital Plot")
plt.savefig("Orbital.png", dpi=300)
```

```python
for i in range(6):
    sim.integrate(sim.t+5)
    sim2.integrate(sim.t+5)
    ob1.update()
    ob2.update()
    ob2.fig.savefig("orbitke_%02d.png"%i)
```
